```python
from matplotlib import pyplot as plt
data['campaign'].plot(kind='hist', bins=20, title='campaign')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(data['housing'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(data, x='campaign', y='housing', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['housing'].value_counts()
    for x_label, grp in data.groupby('default')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('default')
_ = plt.ylabel('housing')
```
